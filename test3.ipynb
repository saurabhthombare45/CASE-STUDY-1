```python
import seaborn as sns
import pandas as pd
# load tips dataset 
tips = sns.load_dataset('flights')
```

```python
flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Convert categorical data to numerical values
tips_numeric = tips.copy()
tips_numeric['sex'] = tips_numeric['sex'].cat.codes
tips_numeric['smoker'] = tips_numeric['smoker'].cat.codes
tips_numeric['day'] = tips_numeric['day'].cat.codes
tips_numeric['time'] = tips_numeric['time'].cat.codes

# Matrix form for correlation data
tips_numeric.corr()
```

```python
sns.heatmap(tips_numeric.corr())
```

```python
sns.heatmap(tips_numeric.corr(), cmap='coolwarm', annot=True)
```
