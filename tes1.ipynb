```python
import seaborn as sns
import pandas as pd
# load tips dataset 
tips = sns.load_dataset('tips')
```

```python
# dispaly first five records 
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='day',palette='coolwarm')
```

```python
sns.barplot(x='sex',y='tip',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, palette='rainbow')
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True, palette="Set1")
sns.swarmplot(x="day", y="total_bill", hue="sex", data=tips, color="k", dodge=True)
```
